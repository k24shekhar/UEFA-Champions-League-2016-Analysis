# Euro2016 DastBoard in python

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings as wr
wr.simplefilter(action='ignore',category=FutureWarning)
```

```python
data=pd.read_csv("Goals.csv",encoding='latin')
data
```

```python
data.head(10)
```

```python
data.columns
```

```python
data_player=data[["full_name","nationality","goals_overall","assists_overall"]]
```

```python
data_player.head(10)
```

```python
dp=data_player[data_player['goals_overall']>2]
dp
```

```python
dp1=data_player[data_player['assists_overall']>1]
dp1
```

```python
data_country=data.groupby('nationality',as_index=False).agg({'goals_overall':'sum','assists_overall':'sum'})
data_country.head(10)
```

```python
dc=data_country[data_country['goals_overall']>5]
dc
```

```python
dc1=data_country[data_country['assists_overall']>3]
dc1
```

```python
f, axes= plt.subplots(2,2,figsize=(15,15))
sns.set(style="whitegrid")

plt.title("Euro2016")

fig1=sns.barplot(x='full_name',y='goals_overall',data=dp, ax = axes[0,0],palette='Spectral')
fig1.set_xticklabels(fig1.get_xticklabels(), rotation=40, ha="right")



fig2=sns.barplot(x='full_name',y='assists_overall',data=dp1, ax = axes[0,1],palette="winter")
fig2.set_xticklabels(fig2.get_xticklabels(), rotation=40, ha="right")

fig3=sns.barplot(x='nationality',y='goals_overall',data=dc, ax = axes[1,0],palette= 'Accent')
fig3.set_xticklabels(fig3.get_xticklabels(), ha="right")


fig4=sns.barplot(x='nationality',y='assists_overall',data=dp1, ax = axes[1,1],palette='coolwarm')
fig4.set_xticklabels(fig4.get_xticklabels(), ha="right")

sns.despine()
plt.show()
```
